```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Top Paying Data Scientist Jobs

```python
df_top_paying_jobs = pd.read_csv('https://raw.githubusercontent.com/RybaFM/Project_data_job_analysis/main/project_sql/result_datasets/top_paying_jobs.csv')
```

```python
df_top_paying_jobs['bar_title'] = df_top_paying_jobs['job_title'] + ', ' + df_top_paying_jobs['company_name']
```

```python
plt.figure(figsize=(14,10))
ax1 = sns.barplot(
    x='salary_year_avg',
    y='bar_title',
    data=df_top_paying_jobs
)
plt.xlabel('Salary(in USD)')
plt.ylabel('')
plt.title('Top Paying Data Scientist Jobs in Central Europe')
ax1.xaxis.grid(True)
ax1.set_axisbelow(True)
plt.savefig('top_paying_jobs_plot.png', bbox_inches='tight')
plt.show()
```

## 5. The Most Optimal Data Scientist Skills

```python
df_optimal_skills = pd.read_csv('https://raw.githubusercontent.com/RybaFM/Project_data_job_analysis/main/project_sql/result_datasets/optimal_skills.csv')
```

```python
ax5 = sns.scatterplot(
    x = 'avg_salary',
    y = 'demand_count',
    hue = 'skills',
    s=50,
    data = df_optimal_skills,
    edgecolor = 'black'
)
plt.xlabel('Average Salary(in USD)')
plt.ylabel('Demand Count')
plt.title('Optimal Data Scientist Skills in Central Europe')

for i, row in df_optimal_skills.iterrows():
    if i % 2 == 0:
      ax5.text(row['avg_salary'], row['demand_count'] + 0.1,
              f"{row['skills']}",
              fontsize=9, alpha=0.8)
    else:
      ax5.text(row['avg_salary'], row['demand_count'] - 0.25,
              f"{row['skills']}",
              fontsize=9, alpha=0.8)

ax5.grid(True)
ax5.set_axisbelow(True)
ax5.legend().remove()
plt.savefig('optimal_skills_plot.png')
plt.show()
```
